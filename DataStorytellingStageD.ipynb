```python
import pandas as pd
production_df = pd.read_csv("Africa Food Production (2004 - 2013).csv", encoding = "latin-1")
```

```python
production_df.describe()
```

```python
production_df.head(20)
```

```python
production_df['Country'].value_counts()
```

```python
production_df['Year'].value_counts()
```

```python
production_df['Item'].value_counts()
```

```python
prod_df_sum = production_df.groupby(["Country","Year"])["Value"].sum()
prod_df_sum.head()
```

```python
import numpy as np

# Normalize the data
plot_data = np.array(prod_df_sum)
#plot_data = (plot_data - np.min(plot_data)) / (np.max(plot_data) - np.min(plot_data))

# Select a specific range of the data
plot_data = plot_data[1:].tolist()
plot_data = plot_data[:4]
```

```python
import matplotlib.pyplot as plt

# Create a logarithmic plot of the data
plt.figure(figsize=(10,6))
plt.semilogy(plot_data)
plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Trend in Food Production (2004-2013)")
plt.show()
```

```python
import matplotlib.pyplot as plt

production_df.groupby(['Year']).sum().plot(y='Value', kind='line', title='Food Production Trend')
plt.xlabel('Year')
plt.ylabel('Production (kt)')
plt.show()
```

Comparison

```python
production_df.groupby(['Year']).sum().plot(y='Value', kind='line', label='Production')
supply_df.groupby(['Year']).sum().plot(y='Value', kind='line', label='Supply')
plt.xlabel('Year')
plt.ylabel('Values (kt)')
plt.title('Food Production vs. Supply')
plt.legend(loc='upper left')
plt.show()
```
