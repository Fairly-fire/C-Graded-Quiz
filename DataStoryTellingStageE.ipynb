```python
#drop the index column
financial_report.drop(columns=['index'], inplace = True)
```

```python
financial_report
```

```python
#defining a function to remove $ and , signs
def replace_func(df, col):
  char_list = ['$ ',',']

  for char in char_list:
      df[col] = df[col].apply(lambda a: str(a).replace(char,''))
```

```python
col_list = ['2009',	'2008',	'2007',	'2006',	'2005']

for col in col_list:
  replace_func(financial_report,col)
```

```python
financial_report
```

```python
#check datatypes
financial_report.dtypes
```

```python
#convert object to float
financial_report[['2009',	'2008',	'2007',	'2006',	'2005']] = financial_report[['2009',	'2008',	'2007',	'2006',	'2005']].astype(float)
```
